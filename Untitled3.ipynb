```python
#import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#seperate daatframes
df_profile=df_lifestyle[['student_id', 'Age', 'Gender', 'Medical Condition']]
df_academic=df_lifestyle[['student_id', 'Academic Performance (GPA)', 'Study Hours Per Week']]
df_clinicalstatus=df_lifestyle[['student_id', 'Mental Stress Level', 'Cognitive Distortions', 'Counseling Attendance', 'Family Mental Health History']]
df_habits=df_lifestyle[['student_id', 'Sleep Duration (Hours per night)', 'Physical Exercise (Hours per week)', 'Diet Quality', 'Social Media Usage (Hours per day)', 'Substance Use']]
df_stress=df_lifestyle[['student_id', 'Financial Stress', 'Peer Pressure', 'Relationship Stress', 'Family Support']]
df_copymech=df_lifestyle[['student_id', 'Stress Coping Mechanisms']]
```

```python
#data cleaning
df_lifestyle = df_lifestyle.dropna()
df_lifestyle = df_lifestyle.drop_duplicates()

df_lifestyle.isnull().sum()

```

```python
#Stress Level Distribution
bins = [0, 3, 6, 10]
labels = ['Low stress (1-3)', 'Moderate stress (4-6)', 'High stress (7-10)']
df_clinicalstatus['Stress Category'] = pd.cut(df_clinicalstatus['Mental Stress Level'], bins=bins, labels=labels, include_lowest=True)
stress_counts=df_clinicalstatus['Stress Category'].value_counts()
```

```python
#evaluating social_media_usage to segregate low,medium,high usage boundaries
df_habits['Social Media Usage (Hours per day)'].value_counts()
```

```python
#Social media usage distribution
# 0-2 hours (Low), 2-5 (Moderate), 5+ (Heavy)
usage_bins = [0, 2, 5, 24]
usage_labels = ['Low User', 'Moderate User', 'Heavy User']
df_lifestyle['Usage Level'] = pd.cut(df_lifestyle['Social Media Usage (Hours per day)'],
                           bins=usage_bins, labels=usage_labels, include_lowest=True)
df_habits['Usage Level'] = pd.cut(df_habits['Social Media Usage (Hours per day)'],
                           bins=usage_bins, labels=usage_labels, include_lowest=True)
```

```python
average_stress_by_social_media = df_lifestyle.groupby('Usage Level')['Mental Stress Level'].mean().reset_index()
# Explicitly set the order for the 'Usage Level' categorical column for clear plotting
category_order = ['Low User', 'Moderate User', 'Heavy User']
average_stress_by_social_media['Usage Level'] = pd.Categorical(average_stress_by_social_media['Usage Level'], categories=category_order, ordered=True)
print("Average Mental Stress Level by Social Media Usage:")
print(average_stress_by_social_media)
```

```python
average_stress_by_social_media.plot(kind='bar', x='Usage Level', y='Mental Stress Level', title='Average Mental Stress Level by Social Media Usage',
                                     xlabel='Social Media Usage Level', ylabel='Average Mental Stbress Level')
plt.ylim(5,7)
plt.show()
```

```python
#sleep duration impact on gpa
import matplotlib.pyplot as plt
sleep_bins = [0, 6, 8,24]
sleep_labels = ['less sleep', 'Moderate sleep', 'good sleep']
df_lifestyle['sleep Level'] = pd.cut(df_lifestyle['Sleep Duration (Hours per night)'],
                           bins=sleep_bins, labels=sleep_labels, include_lowest=True)

average_gpa_by_sleep_level = df_lifestyle.groupby('sleep Level')['Academic Performance (GPA)'].mean().reset_index()

average_gpa_by_sleep_level.plot(kind='bar', x='sleep Level', y='Academic Performance (GPA)',
                                   title='Average Academic Performance (GPA) by Sleep Duration',
                                   xlabel='Sleep Duration Level', ylabel='Average Academic Performance (GPA)')
plt.ylim(1.9, 2.1)
plt.show()
```

```python
#Family Support Effect
#Strong family support is linked with lower mental stress and healthier coping behaviors.
family_support_bins = [0, 1, 3,5]
family_support_labels = ['less support', 'Moderate support', 'full supportive']
df_lifestyle['family support Level'] = pd.cut(df_lifestyle['Family Support'],
                           bins=family_support_bins, labels=family_support_labels, include_lowest=True)

average_mental_stress_by_family_support = df_lifestyle.groupby('family support Level')['Mental Stress Level'].mean().reset_index()

average_mental_stress_by_family_support.plot(kind='bar', x='family support Level', y='Mental Stress Level',
                                   title='Average Mental Stress Level by family suppot',
                                   xlabel='family suppot Level', ylabel='Mental Stress Level')
plt.ylim(5.1,9)
plt.show()
```

```python
average_mental_stress_by_family_support
```

```python
#inancial Stress as a Risk Factor
#Financial stress shows a strong positive correlation with mental stress, independent of academic performance.
financial_bins = [0, 1, 4,5]
financial_labels = ['poor', 'Middle class', 'rich']
df_lifestyle['financial Level'] = pd.cut(df_lifestyle['Financial Stress'],
                           bins=financial_bins, labels=financial_labels, include_lowest=True)

average_mental_stress_by_financial_stress = df_lifestyle.groupby('financial Level')['Mental Stress Level'].mean().reset_index()

average_mental_stress_by_financial_stress.plot(kind='bar', x='financial Level', y='Mental Stress Level',
                                   title='Average Mental Stress Level impact by financial stress',
                                   xlabel='financial Level', ylabel='Mental Stress Level')
plt.ylim(5.2,6)
plt.show()
```
